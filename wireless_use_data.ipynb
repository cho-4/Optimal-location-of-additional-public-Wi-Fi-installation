```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wirelessData = pd.read_csv('../content/sample_data/무선데이터 트래픽 통계.csv', encoding='euc-kr')
# wirelessData.head(100)
# print(wirelessData.dtypes)
# wirelessData = wirelessData.astype({'구분': 'object'})
# print(wirelessData.dtypes)
wirelessData.head(100)
```

```python
plt.rc('font', family='NanumBarunGothic')

plt.figure(figsize=(25, 10))
plt.bar(wirelessData['구분'], wirelessData['합계'], width=0.5)
plt.xlabel('날 짜', fontsize=20)
plt.ylabel('합 계', fontsize=20)
plt.title('무선데이터 트래픽 통계[2023년 1월]', fontsize= 35)
plt.show()
```
