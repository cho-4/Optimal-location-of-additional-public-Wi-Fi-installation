```python
import pandas as pd
from sklearn_extra.cluster import KMedoids
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

# 데이터 불러오기
df = pd.read_csv('../content/sample_data/Variables for Clustering ver2.csv', encoding='euc-kr')

# K-medoids 알고리즘 적용
kmedoids = KMedoids(n_clusters=15, metric='manhattan', random_state=0).fit(df[['위도', '경도']])

# 클러스터별 중심점 좌표 구하기
centers = kmedoids.cluster_centers_
print(centers)

# 클러스터별 데이터 구하기
df['cluster'] = kmedoids.labels_

# 여기서 부턴 특정 X축과 Y축의 범위에서의 시각화 결과를 도출하는 코드
p = sb.lmplot(x='위도', y='경도', data=df, fit_reg=False, scatter_kws={"s": 150}, hue="cluster")

# 클러스터링 결과를 통해 얻은 중심점 좌표 시각화
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=50, alpha=0.5);

plt.title('K-medoids Example')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```
